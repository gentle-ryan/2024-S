```python
# branch
num = 15
if num % 2 == 0:
  print('even number')
else:
  print('odd number')

# nested branches
num = -4
if num > 0:
  if num % 2 == 0:
    print('positive even number')
  else:
    print('positive odd number')
else:
  if num % 2 == 0:
    print('negative even number')
  else:
    print('negative odd number')

# branch with more than 2 cases
age = 22
if age < 13:
  print('child')
elif age >= 13 and age < 20:
  print('teenager')
else:
  print('adult')
```

```python
# factorial
num = 10
result = 1
for i in range(num, 0, -1):
  result = result * i
print(result)

num = 10
result = 1
for i in range(1, num+1, 1):
  result = result * i
print(result)

# factorial with while statement
num = 10
result = 1
while num > 0:
  result = result * num
  num = num - 1
print(result)
```

```python
def print_triangle(n):
  start = n // 2
  end = n // 2
  height = n // 2 + 1
  width = n

  for i in range(0, height):
    for j in range(0, width):
      if start <= j and j <= end:
        print('*', end='')
      else:
        print(' ', end='')
    print('')
    start = start - 1
    end = end + 1
  print('')
```

```python
print_triangle(5)
print_triangle(7)
```
