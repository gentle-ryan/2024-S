현재의 insert fuction은 head에서 시작하여 전체 list를 훑은 뒤에 추가하는 방식이다. 만약 tail node가 있다면, 전체 리스트를 볼 필요없이 리스트의 끝에 새로운 노드를 추가하는 것이 가능할 것이다. 
만약 head가 None이라면, 추가하려는 data인 `new_node`가 head이자 tail이 될 것이다.
만약 head가 None이 아니라면, tail의 next가 `new_node`가 될 것이고 추가한 `new_node`가 새로운 tail이 될 것이다.

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

  def __str__(self):
    return 'Data: ' + str(self.data)
```

```python
class LinkedList:
  def __init__(self):
    self.head = None
    self.tail = None # add a tail node

  def __str__(self):
    msg = ''
    current_node = self.head
    while current_node is not None:
      msg += str(current_node.data)
      if current_node.next is not None:
        msg += ' -> '
      current_node = current_node.next
    if len(msg) == 0:
      msg = 'Empty linked list'
    return msg
  
# Modified insert
  def insert(self, data):
    new_node = Node(data)
    if self.head is None: # check the head is none
      self.head = new_node 
      self.tail = new_node # update the tail
    else: 
      self.tail.next = new_node # new node will be the next of tail
      self.tail = new_node # update the tail

  def search(self, target):
    result = None
    current_node = self.head
    while current_node is not None:
      if current_node.data == target:
        result = current_node
        break
      current_node = current_node.next
    return result

  def delete(self, target):
    prev_node = None
    current_node = self.head
    while current_node is not None:
      if current_node.data == target:
        if current_node == self.head:
          self.head = current_node.next
        else:
          prev_node.next = current_node.next
        break
      prev_node = current_node
      current_node = current_node.next
```

```python
linked_list = LinkedList()
linked_list.insert(12)
linked_list.insert(99)
linked_list.insert(37)
linked_list.insert(24)
linked_list.insert(56)
print(linked_list)

result = linked_list.search(24)
print(result)
print(linked_list.search(100))

linked_list.delete(12)
print(linked_list)
linked_list.delete(37)
print(linked_list)
linked_list.delete(56)
print(linked_list)
linked_list.delete(100)
print(linked_list)
```
