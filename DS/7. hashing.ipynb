```python
# O(n)
def get_majority_element(nums):
  dict_nums = dict()
  for i in nums:
    if i in dict_nums:
      dict_nums[i] += 1
    else:
      dict_nums[i] = 1
  for key in dict_nums:
    if dict_nums[key] > int(len(nums)/2):
      return key

    
```

```python
# O(1)
def get_majority_element(nums):
    majority_element = None
    count = 0

    for num in nums:
        if count == 0:
            majority_element = num
        
        if majority_element == num:
            count += 1
        
        else:
            count -= 1

    return majority_element
```

```python
nums = [2, 2, 1, 1, 3, 2, 2]
print(get_majority_element(nums))
```
